## <p style="font-family:JetBrains Mono; font-weight:normal; letter-spacing:2px; color:#F52800; font-size:70%; text-align:left;padding: 0px;">2. 신고 결과 받기 (정답률 37%)</p>

```python
# id_list = ["con", "ryan"] #* 이용자의 ID가 담긴 문자열 배열
# report = ["ryan con", "ryan con", "ryan con", "ryan con"] #* 각 이용자가 신고한 이용자의 ID 정보가 담긴 문자열 배열 report
# k = 3 #* 정지 기준이 되는 신고 횟수 

id_list = ["muzi", "frodo", "apeach", "neo"]
report = ["muzi frodo","apeach frodo","frodo neo","muzi neo","apeach muzi"]
k = 2
```

```python
def solution(id_list, report, k):
    
    n = len(id_list)
    report = set(report)
    str2idx = {name : i for i, name in enumerate(id_list)}
    baduserList = [0 for i in range(n)] #* 신고 받은 횟수
    reportList = [[] for i in range(n)] #* User가 신고한 사람
    pauseList = []

    for val in report:
        reporter, badUser = val.split() #* 신고자, 신고 받은 ID
        ri, bi = str2idx[reporter], str2idx[badUser] # 신고자 인덱스 , 신고 받은 ID 인덱스
        baduserList[bi] += 1 
        reportList[ri].append(bi) 

    for id, cnt in enumerate(baduserList):
        if cnt >= k:
            pauseList.append(id) 
            
    answer = []
    for id in range(n):
        answer.append(len(set(reportList[id]).intersection(set(pauseList))))
    return answer

solution(id_list, report, k)
```
