```python
def solution(id_list, report, k):
    n = len(id_list)
    str2idx = {name : i for i, name in enumerate(id_list)}
    baduserList = [0 for i in range(n)] #* 신고 받은 횟수
    reportList = [[] for i in range(n)] #* User가 신고한 사람
    pauseList = []
    
    for val in report:
        reporter, badUser = val.split() #* 신고자, 신고 받은 ID
        ri, bi = str2idx[reporter], str2idx[badUser] # 신고자 인덱스 , 신고 받은 ID 인덱스
        baduserList[bi] += 1
        reportList[ri].append(bi) #* 누구를 신고했는지

    for id, cnt in enumerate(baduserList):
        if cnt >= k:
            pauseList.append(id)
    
    answer = []
    for id in range(n):
        answer.append(len(set(reportList[id]).intersection(set(pauseList))))
    
    return answer  

solution(id_list, report, k)
```
